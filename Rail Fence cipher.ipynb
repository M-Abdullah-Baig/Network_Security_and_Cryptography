```python


# Rail Fence cipher implementation
 
# function to encrypt a message
def encrypt(text, key):
 
    # create the matrix to cipher
    # plain text key = rows = rail
    # length(text) = columns

    rail = [['\n' for i in range(len(text))]
                for j in range(key)]
     
    # to find the direction
    dir_down = False
    row, col = 0, 0
     
    for i in range(len(text)):
         
        # check the direction of flow
        # reverse the direction if we've just
        # filled the top or bottom rail
        if (row == 0) or (row == key - 1):
            dir_down = not dir_down
         
        # fill the corresponding alphabet
        rail[row][col] = text[i]
        col += 1
         
        # find the next row using
        # direction flag
        if dir_down:
            row += 1
        else:
            row -= 1
    # now we can construct the cipher
    # using the rail matrix
    result = []
    for i in range(key):
        for j in range(len(text)):
            if rail[i][j] != '\n':
                result.append(rail[i][j])
    return("" . join(result))
     

# function to decrypt a cipher text

def decrypt(cipher, key):
 
    # create the matrix to cipher
    # plain text key = rows = rail
    # length(text) = columns
  
    rail = [['\n' for i in range(len(cipher))]
                for j in range(key)]
     
    # to find the direction
    dir_down = None
    row, col = 0, 0
     
    # mark the places with '_'
    for i in range(len(cipher)):
        if row == 0:
            dir_down = True
        if row == key - 1:
            dir_down = False
         
        # place the underscore 
        rail[row][col] = '_'
        col += 1
         
        # find the next row
        # using direction flag
        if dir_down:
            row += 1
        else:
            row -= 1
             
    # now we can construct the
    # fill the rail matrix
    index = 0
    for i in range(key):
        for j in range(len(cipher)):
            if ((rail[i][j] == '_') and
            (index < len(cipher))):
                rail[i][j] = cipher[index]
                index += 1
         
    # now read the matrix in
    # zig-zag manner to construct
    # the resultant text
    result = []
    row, col = 0, 0
    for i in range(len(cipher)):
         
        # check the direction of flow
        if row == 0:
            dir_down = True
        if row == key-1:
            dir_down = False
             
        # place the underscore
        if (rail[row][col] != '_'):
            result.append(rail[row][col])
            col += 1
             
        # find the next row using
        # direction flag
        if dir_down:
            row += 1
        else:
            row -= 1
    return("".join(result))
 
```

```python
if __name__ == "__main__":
    
    plaintext = str(input("Enter the plain text: "))
    plaintext = plaintext.replace(" ", "")
    
    key = int(input("Enter tye key: "))
    encrypted = encrypt(plaintext, key)
    print("The encrypted text is: ", encrypted)
    
    print("\n")
    
    encrypted = str(input("Enter the decrypted text is: "))
    key = int(input("Enter tye key: "))
    decrypted = decrypt(encrypted, key)
    print("The decrypted text is: ", decrypted)
    
    
```
